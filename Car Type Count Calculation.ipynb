```python
import pandas as pd
import numpy as np
```

```python
get_type_count = pd.read_csv("C:\\Users\\VINIT SANDESH\\Desktop\\DATA ANALYTICS\\PYTHON\\PROJECT DATASETS\\TRAVEL AGENCY.csv")
get_type_count
```

```python
#  Low for values less than or equal to 15,
#  Medium for values greater than 15 and less than or equal to 25,
#  High for values greater than 25.

conditions=[get_type_count['car']<=15,
            (get_type_count['car']>15) & (get_type_count['car']<=25), get_type_count['car']>25]
values = ['low','medium','high']
```

```python
get_type_count['car_type'] = np.select(conditions, values) #Add a new categorical column car_type based on values of the column car:
get_type_count
```

```python
get_type_count['car_type'].value_counts() #Calculate the count of occurrences for each car_type category and return the result as a dictionary. 
```
